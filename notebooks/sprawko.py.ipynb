```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from src.dbscan import DBSCAN
```

```python
X_inliers1, _ = make_blobs(n_samples=100, centers=[[5, 5]], cluster_std=[0.2])
X_outliers1, _ = make_blobs(n_samples=3, centers=[[0, 0]], cluster_std=[0.5])
X_inliers2, _ = make_blobs(n_samples=100, centers=[[5, -5]], cluster_std=[0.2])
X_outliers2, _ = make_blobs(n_samples=3, centers=[[4, -4]], cluster_std=[0.5])
X = np.concatenate([X_inliers1, X_inliers2, X_outliers1, X_outliers2], axis=0)
X = np.random.permutation(X)
```

```python
plt.scatter(X_inliers1[:, 0], X_inliers1[:, 1], color='blue')
plt.scatter(X_inliers2[:, 0], X_inliers2[:, 1], color='blue')
plt.scatter(X_outliers1[:, 0], X_outliers1[:, 1], color='red')
plt.scatter(X_outliers2[:, 0], X_outliers2[:, 1], color='red')

plt.grid()
plt.show()
```

```python
dbscan = DBSCAN()
predictions = dbscan.fit_predict(X)

scatter = plt.scatter(X[:, 0], X[:, 1], c=predictions)

# Get the legend handles and labels
handles, labels = scatter.legend_elements()

# Create custom labels
custom_labels = [f"Cluster n. {i}" for i in range(len(labels))]
custom_labels[0] = "Outliers"

# Create a legend with the custom labels
plt.legend(handles, custom_labels)
plt.grid(True)
plt.show()
```
